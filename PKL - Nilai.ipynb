```
import pandas as pd
import numpy as np
import openpyxl
```

```
pretest=pd.read_excel(r'nilai_pretest-fix.xlsx')
posttest=pd.read_excel(r'nilai_posttest-fix.xlsx')
```

```
pretest
```

```
posttest
```

```
combine = posttest.merge(pretest, on='NIM', how='left')
```

```
combine
```

```
combine = combine.drop(columns=['Nama Mahasiswa_y', 'Mata Kuliah_y'])
combine = combine.rename(columns = {'Nama Mahasiswa_y':'Nama Mahasiswa', 'Mata Kuliah_y':'Mata Kuliah'})
combine
```

```
combine=combine.dropna()
print(combine.shape)
```

```
df_nilai = pd.DataFrame()
df_nilai["NIM"]=combine["NIM"].copy()
df_nilai["nilai_pretest"]=combine["Pre Test"].copy()
df_nilai["nilai_posttest"]=combine["Post Test"].copy()
```

```
df_nilai
```

```
df_nilai=df_nilai.dropna()
df_nilai.describe()
```

```
df_nilai[df_nilai.duplicated(['NIM'], keep=False)]
```

```
df_nilai.drop([245,235,226,195,174,159,162,108,135,136,137,91,11],axis=0, inplace=True)
df_nilai=df_nilai.reset_index(drop=True)
```

```
df_nilai["delta"]=df_nilai["nilai_posttest"]-df_nilai["nilai_pretest"]
df_nilai
```

```
df_nilai.to_csv (r'df_nilai.csv', index = False, header=True)

print (df_nilai)
```
